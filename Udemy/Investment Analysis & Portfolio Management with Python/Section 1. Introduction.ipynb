# IMPORTANT: Pre-Requisites | Please read before enrolling.
We really want to make sure this is clear, so that the expectations of the course are clear, and so you can get the best learning experience.

This is a Finance course - not a Python course.

**That means we've designed this course assuming you know how to code.**

You don't need to be an 'expert' in Python, but...

**/!\ WE ASSUME YOU KNOW:**

- The difference between:
    - "strings", "integers", "floats", and "booleans".
    - "lists", "tuples", and "dictionaries".
- How to write "if else" statements.
- How to "slice" and "concatenate" lists.



In this course, we use Python as a tool to conduct investment analysis and portfolio management.



The emphasis of the course is on Finance.

We dive deep into the equations of Investment Analysis & Portfolio Management. We rip apart each equation one variable at a time, and explain the intuition and rationale behind each and every formula, from scratch.

You'll calculate and work with equations manually first, and only then move on to implementing it on Python.

### THIS COURSE IS NOT FOR YOU IF...

- You want to learn Python from scratch.
- You want to learn Python for Finance from scratch.
- You want to learn about data structures and algorithms.
- You want to learn about how Python works behind the scenes (e.g. how Python manages memory).

### THIS COURSE IS FOR YOU IF...

- You want to learn Investment Analysis & Portfolio Management using Python.
- You want to learn how to quantify and measure the return and risk of individual stocks.
- You want to learn how to quantify and measure portfolio returns and risk.
- You want to learn how to use Python to optimise portfolio returns and risk, minimise portfolio risk, maximise expected returns.

# Accessing Financial Data
- Access to financial data can be quite expensive. Fortunately, there are some free alternatives.

### Accessing Financial Data Manually
- You can access data manually from Yahoo! Finance here: http://finance.yahoo.com
- Alternatively, you can access it from Google Finance by opening a Google Sheets document at http://sheets.google.com/ and using the "GOOGLEFINANCE" function.



### Accessing Financial Data Programatically
Pandas Datareader

- Accessing data via Python's Pandas Datareader has known issues and is not very reliable at this point in time. That is why we do not use it in the course.

Other sources

- There are other reliable, and really useful sources of financial data available via API calls, such that the data can be loaded programatically into Python.
- Unfortunately, we cannot mention those sources nor display them in the course due to copyright restrictions and ambiguity. We spoke with the API providers but did not get a clear, unambiguous response about whether or not we can display them in this course.


### Data in this course
- All of the data in this course is obtained from Yahoo! Finance and / or Google Sheets.
- Throughout the course, you'll have access to the datasets we use in the videos as csv files.
- Note that in the real world, it is very uncommon for people to download data from Yahoo! Finance or Google Sheets when coding in Python.

While we will give you data to work with in the course, we strongly recommend working with your own datasets. This will help you better apply what you learn, and also deal with data cleaning / "data wrangling" effectively.



To find relevant APIs which we cannot name, search "financial and alternative data api python" on Google.
